```python
from gtts import gTTS
import os
from jiwer import wer
import speech_recognition as sr
```

```python
def speech_to_text():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Speak something...")
        recognizer.adjust_for_ambient_noise(source)
        audio = recognizer.listen(source)
    
    try:
        text = recognizer.recognize_google(audio)
        print("Recognized Text:", text)
        return text
    except sr.UnknownValueError:
        print("Could not understand audio")
        return ""
    except sr.RequestError:
        print("Could not request results")
        return ""

def text_to_speech(text, language='en'):
    tts = gTTS(text=text, lang=language, slow=False)
    tts.save("output.mp3")
    os.system("start output.mp3" if os.name == "nt" else "mpg321 output.mp3")

def calculate_accuracy(expected_text, recognized_text):
    error_rate = wer(expected_text, recognized_text)
    accuracy = (1 - error_rate) * 100
    print(f"Model Accuracy: {accuracy:.2f}%")

def main():
    expected_text = input("Enter expected text for accuracy check: ")
    recognized_text = speech_to_text()
    if recognized_text:
        calculate_accuracy(expected_text, recognized_text)
    text_to_speech(recognized_text)

if __name__ == "__main__":
    main()
```
